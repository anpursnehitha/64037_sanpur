```python
#Tokenizing the data
from keras.preprocessing.text import Tokenizer
from keras import preprocessing
import numpy as np

maxlen = 100  # cut reviews after 100 words
training_samples = 200  #  Training on 200 samples
validation_samples = 10000  # Validating on 10000 samples
max_words = 10000  # considering only the top 10,000 words

tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)

word_index = tokenizer.word_index
print('Found %s unique tokens.' % len(word_index))

data = tensorflow.keras.preprocessing.sequence.pad_sequences(sequences, maxlen=maxlen)

labels = np.asarray(labels)
print('Shape of data tensor:', data.shape)
print('Shape of label tensor:', labels.shape)

indices = np.arange(data.shape[0])
np.random.shuffle(indices)
data = data[indices]
labels = labels[indices]

x_train = data[:training_samples]
y_train = labels[:training_samples]
x_val = data[training_samples: training_samples + validation_samples]
y_val = labels[training_samples: training_samples + validation_samples]
```

```python
import os
import tarfile

#importing the drive to use google drive anad mounting it
from google.colab import drive
drive.mount('/content/drive')
#Setting the URL of the zip file that needs to be downloaded
tar_link= 'https://drive.google.com/file/d/1p5XeLhXtkq4WO74rLpZq4IoPQ3nlwddH/view?usp=share_link'
#Extracting the file id from the URL
file_id = tar_link.split('/')[-2]
#Download the file of extracted id
!gdown --id "{file_id}"
#Setting the path to download the file
tar_path = './glove.6B.zip'
#Setting the path where zip file can be extracted
output_folder = '/content/drive/MyDrive/umcompressedglove/'
#Creating output directory
!mkdir -p "{output_folder}"
#Extracting the zip file
!unzip  "{tar_path}" -d "{output_folder}"

glove_dir = '/content/drive/MyDrive/umcompressedglove/'

#Preprocessing the embedding
embeddings_index = {}
f = open(os.path.join(glove_dir, 'glove.6B.100d.txt'))
for line in f:
    values = line.split()
    word = values[0]
    coefs = np.asarray(values[1:], dtype='float32')
    embeddings_index[word] = coefs
f.close()

print('Found %s word vectors.' % len(embeddings_index))

embedding_dim = 100
embedding_matrix = np.zeros((max_words, embedding_dim))
for word, i in word_index.items():
    embedding_vector = embeddings_index.get(word)
    if i < max_words:
        if embedding_vector is not None:
            # Words not found in embedding index will be all-zeros.
            embedding_matrix[i] = embedding_vector
```

```python
# Defining the model
from keras.models import Sequential
from keras.layers import Embedding, Flatten, Dense

model = Sequential()
model.add(Embedding(max_words, embedding_dim, input_length=maxlen))
model.add(Flatten())
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
#Loading GloVe embedding and freezing the model
model.layers[0].set_weights([embedding_matrix])
model.layers[0].trainable = False
```

```python
#Compiling the model
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['acc'])
history = model.fit(x_train, y_train,
                    epochs=10,
                    batch_size=32,
                    validation_data=(x_val, y_val))
model.save_weights('pre_trained_glove_model.h5')
```

```python
# Plotting Training and Validation accuracy 
# Plotting Training and Validation loss
import matplotlib.pyplot as plt

acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(acc) + 1)

plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()

plt.figure()

plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()

plt.show()
```

```python
from keras.models import Sequential
from keras.layers import Flatten, Dense

from keras.datasets import imdb
from keras import preprocessing

# Top Number of common words to consider as features
max_features = 10000
# Cut off review after this number of words
maxlen = 20

# Loading the data as lists of integers.
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=max_features)

# Defining the range of training sample sizes
training_samples_list = [500, 1000, 2500, 5000, 7500, 10000, 15000, 20000, 25000]

val_acc = []
acc=[]
for training_samples in training_samples_list:
    #Limiting the number of training samples
    x_train_subset = x_train[:training_samples]
    y_train_subset = y_train[:training_samples]

# Converting list into a 2D integer tensor of shape `(samples, maxlen)`
    x_train_subset = tensorflow.keras.preprocessing.sequence.pad_sequences(x_train_subset, maxlen=maxlen)
    x_test_subset = tensorflow.keras.preprocessing.sequence.pad_sequences(x_test, maxlen=maxlen)
 #Printing the number of training samples   
    training_size = len(x_train_subset)
    print("Number of training samples:", training_size)
#Building the model
    model = Sequential()
# Specifying the max input length to the Embedding layer
    model.add(Embedding(10000, 8, input_length=maxlen))
# Flattening  3D tensor of embeddings  into a 2D tensor of shape `(samples, maxlen * 8)`
    model.add(Flatten())

#  Compiling the Model
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['acc'])
    model.summary()

# Running the Model
    history = model.fit(x_train_subset, y_train_subset,
                        epochs=10,
                        batch_size=32,
                        validation_split=0.2)
#Appening the validation accuracy and accuracy of the 10th epoch to the empty lists
    val_acc.append(history.history['val_acc'][-1])
    acc.append(history.history['acc'][-1])
```

```python
# Plotting Validation accuracy and Training sample size 
# Plotting Accuracy and Training sample size
plt.plot(training_samples_list, acc, 'bo-',color='red', label=' Accuracy')
plt.plot(training_samples_list, val_acc, 'bo-', color='blue', label='Validation Accuracy')
plt.title('Validation Accuracy and Accuracy vs. Number of Training Samples')
plt.xlabel('Training Samples')
plt.ylabel('Performance')
plt.legend()
plt.show()
```
